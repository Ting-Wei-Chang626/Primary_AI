```python
import csv
```

```python
animals = ["dog", "rooster", "pig", "cow", "frog",
           "cat", "hen", "insects", "sheep", "crow"]
data = []
data_animal = []
with open('/content/drive/MyDrive/NTUCE_AI_worshop_0629/ESC-50/meta/esc50.csv', newline='') as csvfile:

    rows = csv.reader(csvfile)
    for row in rows:
        data.append(row)

data_animal.append(data[0])
for row in data:
    if row[3] in animals:
        data_animal.append(row)


with open('/content/drive/MyDrive/NTUCE_AI_worshop_0629/ESC-50/meta/output.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerows(data_animal)
```
